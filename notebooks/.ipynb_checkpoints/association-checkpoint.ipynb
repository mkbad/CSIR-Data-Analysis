# Create CSVs for association

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../main_data_id.csv", index_col = 'id')
```

```python
crop_harvest_df = pd.read_csv("../data/crop_harvest.csv")
```

```python
crop_harvest_df.head()
```

```python
def get_id_by_value(value,dataframe):
    value = value.strip()
    res = dataframe[dataframe["Name"]==value].values[0][0]
    return res
```

```python
#The error message suggests that you have missing or NaN values in the 
#'crop_harvest' column of the dataframe
```

```python
#fixing the error
for ind in df.index:
    row_id = df['record_id'][ind]
    values = df['crop_harvest'][ind]
    if pd.notna(values):  # Check if the value is not NaN
        list_of_values = values.split(",")
        for item in list_of_values:
            print(f"Record ID: {row_id}, Crop Harvest data: {item}")
    else:
        print(f"Record ID: {row_id}, Crop: No crop data available")
```

```python
# Create a new DataFrame with 'record_id' and 'crop_harvest' columns
assoc_df = df[['record_id', 'crop_harvest']]

# Save the new DataFrame to a CSV file
assoc_df.to_csv("../data/crop_harvest_association_trial.csv", index=False)
```

```python
#value mapping to 'crop_harvest' ID
```

```python
value_id_mapping = dict(zip(crop_harvest_df['Name'], crop_harvest_df['ID']))
```

```python
df['Associated_ID'] = df['crop_harvest'].apply(lambda x: [value_id_mapping[val.strip()] for val in str(x).split(',') if val.strip() in value_id_mapping])
```
