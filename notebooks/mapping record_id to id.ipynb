```python
import pandas as pd
```

```python
maindf = pd.read_csv('../md.csv')
maindf.head()
```

```python
maindf.columns
```

```python
extracted_df = pd.read_csv('../data/extracted/type_of_cement_grade.csv')
extracted_df.head()
```

```python
association = []
```

```python
for ind in maindf.index:
    row_id = maindf.loc[ind, 'record id']
    values = maindf.loc[ind, 'type of cement grade']
    if pd.notna(values):
        list_of_values = values.split(",")
        for value in list_of_values:
            matching_ids = extracted_df.loc[extracted_df['value'] == value.strip(), 'id'].values
            if len(matching_ids) > 0:
                association_id = matching_ids[0]
                association.append((row_id, association_id))
```

```python
association_df = pd.DataFrame(association, columns=['record id', 'id'])
```

```python
association_df
```

```python
association_df.to_csv('../data/association/type_of_cement_grade-association.csv', index=False)
```
